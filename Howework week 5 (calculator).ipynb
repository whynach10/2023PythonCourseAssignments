```python
#define a function for variable messages
def give_message(firstnumber:float, secondnumber:float, result:float, operand:str)->str:
    message = f"{firstnumber} {operand} {secondnumber} is equal to {result}"
    return (message)
#defien a function for variable 
def inform_user(operation:str) -> str:
    """
    This function informs the user
    """

    message= f"""
    The entered numbers will be {operation}

    """
    return message

def addition (firstnumber, secondnumber):
    print(inform_user("added"))
    result = firstnumber + secondnumber
    return(result)

def substraction (firstnumber, secondnumber):
    print(inform_user("substracted"))
    result = firstnumber - secondnumber
    return(result)
def multiplication (firstnumber, secondnumber):
    print(inform_user("multiplied"))
    result = firstnumber * secondnumber
    return(result)
def divison (firstnumber, secondnumber):
    print(inform_user("divided"))
    result = firstnumber / secondnumber
    return(result)

#This is the menu choice that is presented to the user

print("""
xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx
Please choose your desired calculation by typing the number infront of it. By typing q the program will quit
xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx
1. Addition
2. Substraction
3. Multiplication
4. Division
q. Quits the menu

xxxxxxxxxxxxxxxxxxxxxxxxxxxxxx

""",flush = True) #flush is true to make sure the message appears first

#User is asked to enter a number to determine which conversion funtion should be called
while True:
    menu_choice = input("Please enter the operation you wanna calculate with from the menu. Only type the number! Or type q to quit the program")
#In case the user entered a capital "Q", make it lowercase    
    menu_choice = menu_choice.lower()
#Fist a check if the program should quit
    if menu_choice == "q":
        print("program stops!")
        break
#User input control: was a number entered?        
    elif menu_choice.isdigit() == True:  
#The user is prompted to give 2 numbers. Depending on the calculation choice above the corrseponding function is callled (addition, substraction, multiplication, divison)
#And the user is presented with the calculated result
        if menu_choice == "1":
            firstnumber = float(input("Please enter the first number you want to calculate with:"))
            secondnumber = float(input("Please enter the second number you want to calculate with:"))
            result = addition(firstnumber, secondnumber)
            print (give_message(firstnumber, secondnumber, result, "+"))
        elif menu_choice == "2":
            firstnumber = float(input("Please enter the first number you want to calculate with:"))
            secondnumber = float(input("Please enter the second number you want to calculate with:"))
            result = substraction(firstnumber, secondnumber)
            print (give_message(firstnumber, secondnumber, result, "-"))
        elif menu_choice == "3":
            firstnumber = float(input("Please enter the first number you want to calculate with:"))
            secondnumber = float(input("Please enter the second number you want to calculate with:"))
            result = multiplication(firstnumber, secondnumber)
            print (give_message(firstnumber, secondnumber, result, "*"))
        elif menu_choice == "4":
            firstnumber = float(input("Please enter the first number you want to calculate with:"))
            secondnumber = float(input("Please enter the second number you want to calculate with:"))
            result = divison(firstnumber, secondnumber)
            print (give_message(firstnumber, secondnumber, result, "/"))
#User input control: Was the number entered bigger than the choice of numbers on the menu?
        elif int(menu_choice) >= 7:
            print ("please enter a number between 1 and 4 or q to stop the program")
#The final user input control in case the user entered a non-number (that is not "q")
        else:
            print("please enter a number between 1 and 4 or q to stop the program")
        


```

```python
#define a function for variable messages
def give_message(user_equotation:str, result:float)->str:
    message = f"{user_equotation} is equal to {result}"
    return (message)
#this function breaks down the equotation the user entered to be able to calculate it. Then it calculates it
def solve_equotation(user_equotation:str, operand_position:int):
    operand = user_equotation[operand_position]
    firstnumber = float(user_equotation[0:operand_position])
    secondnumber = float(user_equotation[operand_position+1:len(user_equotation)])
    if operand == "+":
        result = firstnumber + secondnumber
    elif operand == "-":
        result = firstnumber - secondnumber
    elif operand == "*":
        result = firstnumber * secondnumber    
    elif operand == "/":
        result = firstnumber / secondnumber    
    return (result)


#The user is asked to enter an equotation
while True:
    user_equotation = input("Please enter a formula in the form of (a) (operant) (b) where a and b are numbers and operant can be +, -, *, /. Enter q to stop the program")
    #In case the user entered a capital "Q", make it lowercase    
    user_equotation = user_equotation.lower()
    #Fist a check if the program should quit
    if user_equotation == "q":
        print("program stops!")
        break
#determine which operant was used and run the funtion to solve. User is presented with a message of the result.
    elif user_equotation.find("+") > 0:
        operand_position = user_equotation.find("+")
        result = solve_equotation(user_equotation, operand_position)
        print (give_message(user_equotation, result))
    elif user_equotation.find("-") > 0:
        operand_position = user_equotation.find("-")
        result = solve_equotation(user_equotation, operand_position)
        print (give_message(user_equotation, result))
    elif user_equotation.find("*") > 0:
        operand_position = user_equotation.find("*")
        result = solve_equotation(user_equotation, operand_position)
        print (give_message(user_equotation, result))
    elif user_equotation.find("/") > 0:
        operand_position = user_equotation.find("/")
        result = solve_equotation(user_equotation, operand_position)
        print (give_message(user_equotation, result))

#The final user input control in case the user entered a non-number (that is not "q")
    else:
        print("The equotation you entered could not be calculated. Please check the synthax. Please only use +,-, * or / operators and only one!")
        
```
